```python
import numpy as np
import matplotlib.pyplot as plt

# Método de Euler implícito con Newton-Raphson
def euler_implicito(u0, t0, tf, N):
    h = (tf - t0) / N
    t = np.linspace(t0, tf, N+1)
    u = np.zeros(N+1)
    u[0] = u0
    for n in range(N):
        x = u[n]  # estimación inicial
        for _ in range(10):  # iteraciones de Newton
            x -= (x - u[n] - h*np.sin(x)) / (1 - h*np.cos(x))
        u[n+1] = x
    return t, u

# Parámetros del problema
u0, t0, tf = 1, 0, 1
Ns = [20, 30, 50, 100]

# Graficar cada aproximación por separado
for N in Ns:
    tN, uN = euler_implicito(u0, t0, tf, N)
    plt.figure()
    plt.plot(tN, uN, 'o-', label=f"N={N}")
    plt.xlabel("t")
    plt.ylabel("u(t)")
    plt.title(f"Aproximación Euler implícito — N={N}")
    plt.legend()
    plt.grid(True)
    plt.show()
```
