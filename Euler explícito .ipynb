```python
import numpy as np
import matplotlib.pyplot as plt

# Solución exacta
def u_exacta(t):
    K = np.tan(0.5)  # tan(1/2) porque u(0) = 1
    return 2 * np.arctan(K * np.exp(t))

# Método de Euler explícito
def euler(N, T=1, u0=1):
    h = T / N
    t = np.linspace(0, T, N+1)
    u = np.zeros(N+1)
    u[0] = u0
    for i in range(N):
        u[i+1] = u[i] + h * np.sin(u[i])
    return t, u

# Valores de N que vamos a probar
Ns = [20, 30, 50, 100]

# Valor exacto en t=1
u_real_1 = u_exacta(1)
print(f"Valor exacto en t=1: {u_real_1:.6f}\n")

# Recorremos cada N y mostramos resultados + gráfica
for N in Ns:
    t, u = euler(N)
    error = abs(u[-1] - u_real_1)
    print(f"N={N:3d}, h={1/N:.4f}, u(1) aprox={u[-1]:.6f}, error={error:.6e}")

    # Gráfica para este N
    t_fino = np.linspace(0, 1, 500)
    plt.figure()
    plt.plot(t_fino, u_exacta(t_fino), 'k-', label="Exacta")
    plt.plot(t, u, 'ro-', label=f"Euler N={N}")
    plt.xlabel("t")
    plt.ylabel("u(t)")
    plt.title(f"Aproximación con Euler explícito (N={N})")
    plt.legend()
    plt.grid(True)
    plt.show()
```
