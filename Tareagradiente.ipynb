```python
import matplotlib.pyplot as plt


precios = [0.59, 0.80, 0.95, 0.45, 0.79, 0.99, 0.90, 0.65, 0.79, 0.69, 0.79,
           0.49, 1.09, 0.95, 0.79, 0.65, 0.45, 0.60, 0.89, 0.79, 0.99, 0.85]

ventas = [3980, 2200, 1850, 6100, 2100, 1700, 2000, 4200, 2440, 3300, 2300,
          6000, 1190, 1960, 2760, 4330, 6960, 4160, 1990, 2860, 1920, 2160]



alpha = 0.01
epocas = 20000


B = 0.0
A = 0.0

m = len(ventas)
historial_costo = []




for _ in range(epocas):

    y_pred = [B + A * xi for xi in precios]


    errores = [yp - yi for yp, yi in zip(y_pred, ventas)]


    dB = sum(errores) / m
    dA = sum(ei * xi for ei, xi in zip(errores, precios)) / m


    B -= alpha * dB
    A -= alpha * dA


    costo = sum(ei**2 for ei in errores) / (2*m)
    historial_costo.append(costo)

print("Par√°metros encontrados:")
print(f"Intercepto (B): {B}")
print(f"Pendiente (A): {A}")


plt.scatter(precios, ventas, color="blue", label="Datos reales")


y_linea = [B + A * xi for xi in precios]
plt.plot(precios, y_linea, color="red", label="Recta ajustada")

plt.xlabel("Precio ($)")
plt.ylabel("Ventas por semana")
plt.legend()
plt.show()


plt.plot(range(epocas), historial_costo)
plt.xlabel("Iteraciones")
plt.ylabel("Costo (ECM)")
plt.title("Convergencia del gradiente descendente")
plt.show()
```
